### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

x = list(range(11)) 
y = [i**2 for i in x]  

plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')

plt.title('Square Function Plot')
plt.xlabel('x values')
plt.ylabel('y = x^2')
plt.grid(True)  
plt.legend()  

plt.show()
```
